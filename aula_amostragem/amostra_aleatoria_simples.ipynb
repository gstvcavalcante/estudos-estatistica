# POPULAÇÃO E AMOSTRA

```python
#População entende-se como todos os dados que compõem o dataset, enquanto a amostra (sample) seria apenas uma parcela dos dados
#que retiramos para realizar a análise.
#No exemplo de uma pesquisa, a população seria propriamente todos os habitantes do país, enquanto a amostra seriam a quantidade
#de pessoas escolhidas para responder a pesquisa.
#A amostra sempre deve ser randômica e representativa, nunca enviesada.
```

### Amostra aleatória simples

```python
#Para retirar uma amostragem aleatória simples com o Python, podemos usar funções prontas de bibliotecas para essa finalidade.

#Importando as bibliotecas 'pandas' para análise e manipulação de dados, 'numpy' para realizar operações e cálculos e 'random'
#que possui funções randômicas dentro do próprio python.

#Ao importar uma biblioteca e usar o 'as' é possível dar um apelido(alias) para facilitar a codificação e leitura.

import pandas as pd
import numpy as np
import random
```

```python
#Aqui utilizei a função do pandas para ler um arquivo csv contendo as informações do census dos Estados Unidos e inseri na
#variável 'dataset'. Dessa forma, a variável passa a conter uma 'tabela de dados', um dataframe, como se fosse uma planilha
#de excel ou uma tabela de SQL.

dataset = pd.read_csv("census.csv")
```

```python
#Essa função shape do pandas retorna a quantidade de linhas e colunas de um dataframe. (syntax = nome_do_dataframe.shape)

dataset.shape
```

```python
#A função head() traz as cinco primeiras linhas do dataframe para que você possa se familiarizar e entender quais dados contidos
#nas colunas.

dataset.head()
```

```python
#Assim como a a head(), agora a função tail() traz as cinco últimas linhas do dataframe. É interessante verificar para garantir
#que todas as linhas seguem o padrão da tabela. A última linha pode trazer campos de valores totais somados por exemplo, como
#ocorre em alguns arquivos de excel.

dataset.tail()
```

```python
#Retirar uma amostra simples:

df_amostra_aleatoria_simples = dataset.sample(n = 100)

#Com esse comando, criamos uma nova variável para abrigar um dataframe com a amostra de nossa população.
#A função sample() faz parte da bilbioteca 'random'  e obtém registros aleatórios do nosso dataset para inserir na nova variável
#Importante notar que a função recebe alguns parâmetros. (n = 100) seria a quantidade de registros aleatórios que contemplarão
#nossa amostragem.
```

```python
#Utilizando novamente os comandos shape e head() podemos notar que o dataframe está composto exatamente por 100 registros,
#E, conforme esperado, são todos aleatórios.
df_amostra_aleatoria_simples.shape
```

```python
df_amostra_aleatoria_simples.head()
```

```python
#Existe um outro parâmetro dentro da função sample() que é o random_state = 'n'. Com esse parâmetro, você pode atribuir aqueles
#exatos registros aleatórios que foram gerados ao número que você inserir no random_state, e sempre que aquele número for 
#inserido como parâmetro, a função retornará os mesmos registros.
df_amostra_aleatoria_simples = dataset.sample(n = 100, random_state = 1)
df_amostra_aleatoria_simples.head()

```

```python
df_amostra_aleatoria_simples = dataset.sample(n = 100, random_state = 2)
df_amostra_aleatoria_simples.head()
```

```python
df_amostra_aleatoria_simples = dataset.sample(n = 100, random_state = 1)
df_amostra_aleatoria_simples.head()
```

```python
#E para melhor entendimento, é possível identificar acima que quando o parâmetro está como 1, o dataframe possui determinados 
#registros, e quando alteramos o parâmetro do random_state para 2, novos registros aleatórios são inseridos no dataframe e 
#ficam vinculados aquele parâmetro.
```

```python
#Uma função pode ser criada dentro do Python para gerar uma amostragem simples dentro de uma variável à partir de um dataframe.
#Definimos o nome da função como 'amostragem_aleatoria_simples' e ela recebe dois parâmetros: um dataframe e a quantidade de 
#amostras que queremos que sejam geradas à partir da nossa população.
#O retorno da função é justamente a função sample().

def amostragem_aleatoria_simples(dataframe, amostras):
    return dataframe.sample(n = amostras)
```

```python
#Agora vamos utilizar a função para inserir as amostras dentro de uma variável.
df_amostra_funcao_aleatoria_simples = amostragem_aleatoria_simples(dataset, 150)
df_amostra_funcao_aleatoria_simples.shape
```

```python
#Acima é possível identificar que criamos uma nova variável, atribuímos dentro dela o retorno da função utilizando aquele nosso
#dataset inicial e uma amostra aleatória de 150 registros.
#Vamos usar o head() mais uma vez para conferir:
df_amostra_funcao_aleatoria_simples.head()
```
